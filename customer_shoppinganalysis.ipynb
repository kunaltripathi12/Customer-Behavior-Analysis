## Analyze Null Value

```python
# Size, Purchase Amount (USD) , Review Rating  , Previous Purchases  

df['Size'].unique()
```

```python
df.groupby('Category')['Size'].unique()
```

```python
df.groupby('Item Purchased')['Size'].unique()
```

```python
# Replacing null & vaules in Electronics category
df.loc[df['Category']=='Electronics','Size']="Not Applicable"
```

```python
df.groupby('Category')['Size'].unique()
```

```python
df.loc[df['Category']=='Accessories','Size']="Free Size"
```

```python
df.groupby('Category')['Size'].unique()
```

```python
#Replacing null & vaules in Footwear category
df.loc[df['Category']=='Footwear','Size']="Not Applicable"
```

```python
df.groupby('Category')['Size'].unique()
```

```python
# Replace null values in Clothing category
df.loc[(df["Category"]=="Clothing") & (df["Size"].isnull()), "Size"] = df[df["Category"]=="Clothing"]["Size"].mode()[0]
```

```python
df.groupby('Category')['Size'].unique()
```

```python
df.groupby("Item Purchased")["Review Rating"].mean()
```

```python
#filling null values of review rating with the mean of each item Purchased
df['Review Rating']=df['Review Rating'].fillna(df.groupby('Item Purchased')['Review Rating'].transform('mean'))
```

```python
df.isnull().sum()
```

```python
df['Review Rating']=df['Review Rating'].round(2)
```

## Previous Purchased

```python
#filling previous Purchased with Zero(0)
df['Previous Purchases']=df['Previous Purchases'].fillna(0)
```

```python
df.isnull().sum()
```

## Purchase Amount

```python
df.groupby("Item Purchased")['Purchase Amount (USD)'].mean()
```

```python
#filling null values of Purchase Amount (USD) with the mean of each item Purchased
df['Purchase Amount (USD)']=df['Purchase Amount (USD)'].fillna(df.groupby('Item Purchased')['Purchase Amount (USD)'].transform('mean'))
```

```python
df.isnull().sum()
```

```python
from sqlalchemy import create_engine
import urllib

# Your SQL Server details
server = r"KT\SQLEXPRESS"      # Use raw string for backslash
database = "customer_analysis"   # Your database name
driver = "ODBC Driver 17 for SQL Server"   # Must be installed on your machine

# Build connection string for Windows Authentication
connection_string = urllib.parse.quote_plus(
    f"DRIVER={{{driver}}};"
    f"SERVER={server};"
    f"DATABASE={database};"
    f"Trusted_Connection=yes;"
)

# Create engine
engine = create_engine(f"mssql+pyodbc:///?odbc_connect={connection_string}")

print("✅ Successfully connected to SQL Server (Windows Authentication)!")
```

```python
table_name = "Customer_behaviour_analysis" #table name
df.to_sql(table_name, engine, if_exists="replace", index=False)

print(F" Data Successfully loaded into table '(table_name)' in database 'customer_analysis'.")
```
